```python
# Import modules
from abtem import *
from abtem.structures import orthogonalize_cell
from ase import Atoms
from ase.io import read
from ase.build import surface
from abtem.visualize import show_atoms
from abtem.measure import Measurement
import matplotlib.pyplot as plt
from abtem.transfer import CTF
```

## Reading the cif or xyz file

```python
# cif_file = read('2H_MoSe2.cif')  # abtem can read cif file and xyz file that has the atomic co-ordinates information
xyz_file = read('WTe2-Td-2L_rep6_10_1.xyz')
title = 'WTe2-Td-2L_rep6_10_1'
# Need to orthogonalize cell
# structure = orthogonalize_cell(cif_file* (15, 10, 1))  # abTEM requires orthogonal cells, thus the orthogonalize function is used
structure = orthogonalize_cell(xyz_file)

# Displaying the structure
show_atoms(structure,
            plane = 'xy', # Look at the xy plane
            legend = True) # Display legend
plt.title(f'{title}')
```

## Adding thermal phonons

```python
#Center the structure on the cell and add a vacuum gap on the z axis
structure.center(
            axis = 2, # Axis to add the vacuum gap
            vacuum = 3 # Vacuum gap of 3 A
        )

# Create phonon configurations using FrozenPhonons
structure_phonons = FrozenPhonons(
            structure, # Strucure to apply phonons
            num_configs = 10, # Number of phonon configuration
            sigmas = {'W': 0.023, 'Te': 0.1},  # Dictionary with displazement due to phonons (in Armstrom) for all the species
            directions = 'xy' # Direction of the displacement
        )

show_atoms(next(iter(structure_phonons)), #Display one of the phonon configuration
                   plane = 'xy', # Look at the xy plane
                   legend = True) # Display legend
plt.title(f'{title}')
```

## Defining potential

```python
tds_potential = Potential(
            structure_phonons, # Desired structure
            sampling = 0.05, # Lateral sampling of the potential in 1/A
            gpts = 1200, # Number of grid points describing each slice (it influences the maximum scattering angle)
            slice_thickness = 1, # Thickness of the slices in A. This will determine the amount of slices
        # The projection is how the 3D potential will be integrated, it can be ither 'infinite' or 'finite'
        # 'infinite' is chosen as it is the least computationaly expensive of the methods
            projection = 'infinite',
        # Potential parametrization describing the radial dependece of the potential for each element
        # Methods available are 'kirkland' or 'lobato'
            parametrization = 'kirkland',
            precalculate = True, # Precalculates the potential, and stores it
            device = 'gpu' # Device to run the simulations, only useful is a NVIDIA GPU with Cuda is available
            #storage = 'cpu' # Place to store the potentials for intensive calculations using GPU
     )
print(f"Real space sampling: {tds_potential.sampling} Angstrom ")
    # Calculate the potentials
    # If progress of calculation is deisred set pbar=True

# Project the sum of all the potentials
tds_potential.project().show(figsize=(4,4))
plt.title(f'{title}')
```

## STEM simulation with PRISM

```python
# Defining the  contrast transfer function
c30 = -0.0005e7
c50 = 0.5e7

# Defining the parameters
parameters = {'C30': c30, 'C50': c50}
aberrations = CTF(parameters=parameters, energy=80e3)

# Defining the scattering matrix
S = SMatrix(expansion_cutoff=25, # Maximum angle for the convergence angle (Should be slightly bigger than probe convergence angle)
            interpolation=2, # By increasing the interpolation we restrict the extent of the probe window, decreasing cost
            energy=80e3, # electron beam energy 300 keV
            semiangle_cutoff=21.3, # Probe convergence angle (mrad)
            # gpts=256,     # is a duplicate to potential
            ctf=aberrations,
            defocus=20,
            device = 'gpu' # Device to run the simulations, only useful is a NVIDIA GPU with Cuda is available
            #storage = 'cpu' # Place to store the potentials for intensive calculations using GPU
           )

# Printing the maximum scattering angle with that potential
S.grid.match(tds_potential)
print(S.cutoff_scattering_angles)

# Create the area for imaging
gridscan = GridScan(
        start=[7, 4], # Starting position for the scan
        end = [36, 33], # Final position for the scan
        sampling = 0.05 # Sampling rate of scan positions [1 / Å]
)


ax, im = tds_potential.project().show()
# Overlay the gridscan
gridscan.add_to_mpl_plot(ax)
```

## Detectors and scan

```python
# Defining the detectors
# Create HAADF, BF, and ABF detectors
haadf = AnnularDetector(inner=51, outer=300, save_file=f'{title}_haadf.hdf5')
#  bf = AnnularDetector(inner=0, outer=10, save_file=f'{title}_bf.hdf5')
#  abf = AnnularDetector(inner=10, outer=40, save_file=f'{title}_abf.hdf5')

# Start the scan for each detector
# The scan function needs the area of interest (gridscan), detector settings, and the structure potential
haadf_scans = S.scan(gridscan, haadf, tds_potential)
#     bf_scans[key] = S.scan(gridscan, bf, tds_potential)
#     abf_scans[key] = S.scan(gridscan, abf, tds_potential)

print(f'The run for {title} finished')
```

```python
# Saving the measurements
measurement = Measurement.read(haadf_scans) # Read the calculated file
# noisy = poisson_noise(measurement, 1*10**7)  # If you want to add poisson noise to the simulated structure
# image_haadf = noisy
image_haadf= measurement  # Comment it out if noise is required
image_haadf.show(title=f"{title}_haadf")
image_haadf.save_as_image(f'{title}.tif')
```
